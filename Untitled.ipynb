```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# Create DataFrame with example vectors
data = {
    'Vector1': [1, 2, 3],
    'Vector2': [4, 5, 6],
    'Vector3': [7, 8, 9],
    'Vector4': [1, 0, 0],
}

df = pd.DataFrame(data).T
df.columns = ['x', 'y', 'z']

# Calculate cosine similarity
cos_sim_matrix = cosine_similarity(df)
cos_sim_df = pd.DataFrame(cos_sim_matrix, index=df.index, columns=df.index)

# Print cosine similarity matrix
print("Cosine Similarity Matrix:\n", cos_sim_df)

# Visualize the vectors
plt.figure(figsize=(10, 6))
for i, (name, vector) in enumerate(df.iterrows()):
    plt.quiver(0, 0, vector['x'], vector['y'], angles='xy', scale_units='xy', scale=1, label=name)

plt.xlim(-1, 8)
plt.ylim(-1, 10)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.title('Vector Representation in 2D Space')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from mpl_toolkits.mplot3d import Axes3D  # Import 3D toolkit

# Create DataFrame with example vectors
data = {
    'Vector1': [1, 2, 3],
    'Vector2': [4, 5, 6],
    'Vector3': [7, 8, 9],
    'Vector4': [1, 0, 0],
}

df = pd.DataFrame(data).T
df.columns = ['x', 'y', 'z']

# Calculate cosine similarity
cos_sim_matrix = cosine_similarity(df)
cos_sim_df = pd.DataFrame(cos_sim_matrix, index=df.index, columns=df.index)

# Print cosine similarity matrix
print("Cosine Similarity Matrix:\n", cos_sim_df)

# Visualize the vectors in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for i, (name, vector) in enumerate(df.iterrows()):
    ax.quiver(0, 0, 0, vector['x'], vector['y'], vector['z'], 
              arrow_length_ratio=0.1, label=name)

# Set limits
ax.set_xlim([0, 8])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])

# Set labels and title
ax.set_title('3D Vector Representation')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('RoomMatesDataSet.csv')
attributes = [
    "Cleanliness Level", "Sleeping Level", "Cooking Level",
    "Introvert Level", "Extrovert Level", "Intrest In Music",
    "Intrest In Dance", "Intrest In Physical Activities",
    "Exercise Habits", "Intrest in Gaming"
]

# Create a bar plot for each individual
fig, ax = plt.subplots(figsize=(10, 6))
df.sample(n=5).set_index('Name')[attributes].plot(kind='bar', ax=ax)
plt.title('Interest Levels by Individual')
plt.ylabel('Level')
plt.xlabel('Names')
plt.xticks(rotation=90)
plt.legend(title='Attributes')
plt.tight_layout()
plt.show()
```
