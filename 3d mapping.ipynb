```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load your dataset
df = pd.read_csv('RoomMatesDataSet.csv')

# Select three features for the 3D plot
x = df['Cleanliness Level']
y = df['Smoking']
z = df['Cooking Level']

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(x, y, z)

# Label the axes
ax.set_xlabel('Cleanliness Level')
ax.set_ylabel('Smoking')
ax.set_zlabel('Cooking Level')

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load your dataset
df = pd.read_csv('RoomMatesDataSet.csv')

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='Cleanliness Level', y='Smoking', z='Cooking Level', 
                    color='Gender',  # Optional: Color points by another feature
                    title='3D Scatter Plot of Roommate Preferences')

# Show the plot
fig.show()
```
